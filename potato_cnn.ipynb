```python
import os
import tensorflow as tf 
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential, Model 
from tensorflow.keras.layers import Dense, Flatten, Dropout, MaxPooling2D
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen= ImageDataGenerator(rescale=1./255, zoom_range=0.2)  
val_datagen= ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    "/kaggle/input/potato-leaf-disease-dataset",
    target_size=(256,256),
    batch_size=32, 
)
```

```python
val_generator = val_datagen.flow_from_directory(
    "/kaggle/input/potato-leaf-disease-dataset",
    target_size=(256,256),
    batch_size=32, 
)
```

```python
from tensorflow.keras import datasets, layers, models
```

```python
model = models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3,3),activation='relu',input_shape=(256,256,3)),
    layers.MaxPooling2D((2,2)), 

    layers.Conv2D(filters=64, kernel_size=(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(filters=128, kernel_size=(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid') 
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    epochs=2,
    validation_data=val_generator
)
```

```python
plt.figure()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('Accuracy')

plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('Loss')

plt.show()
```

```python
model.save("potato.keras")
```
